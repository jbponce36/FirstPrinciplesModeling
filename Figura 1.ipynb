```python
#Figura 1. Serie de tiempo de individuos infectados observados en Gambia y Finlandia.

import numpy as np
import matplotlib.pyplot as plt


#Datos Gambia
g1 = np.loadtxt('gambia.txt')
g2 = np.loadtxt('gambiaweather.txt')
ng1 = len(g1)
tg1 = np.linspace(0,ng1/12,ng1)
ng2 = len(g2)
tg2 = np.linspace(0,ng2/12,ng2)                
           
#Grafico Gambia
plt.figure(figsize=(12,4))
plt.plot(0.5+tg1,g1,'-x',label='Prevalencia de infecciosos')
plt.plot(0.5+tg2,g2,'-o',label='Temperatura promedio mensual')
plt.xlabel('Tiempo (años)')
plt.title('Gambia')
plt.legend(loc=9)


#Datos Finlandia
f1 = np.loadtxt('finland.txt')
f2 = np.loadtxt('finlandweather.txt')
nf1 = len(f1)
tf1 = np.linspace(0,nf1/12,nf1)
nf2 = len(f2)
tf2 = np.linspace(0,nf2/12,nf2) 

#Grafico Finlandia
plt.figure(figsize=(12,4))
plt.plot(tf1,f1,'-x',label='Prevalencia de infecciosos')
plt.plot(tf2,f2,'-o',label='Temperatura promedio mensual')
plt.xlabel('Tiempo (años)')
plt.title('Finlandia')
plt.legend(loc=9)

plt.show()
```
